```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df=pd.read_csv("iris.data")
df
```

```python
df.columns=['sepallength','sepalwidth', 'petallength','petalwidth','species']
```

```python

df
```

```python
df.iloc[5:11].plot(kind='bar',figsize=(20,10))
```

```python
df.plot(figsize=(20,8))
```

```python
df.iloc[[5]].plot(kind='bar',figsize=(20,9))
```

```python
df.iloc[[5]].plot(kind='barh',figsize=(20,9))
```

```python
df['sepalwidth'].plot(kind='hist')   #frequency means how many dataserts are available between a particular bins
```

```python
df.hist(figsize=(20,9), color='g', alpha=.2) #alpha tella us the opacity of the histogram
```

```python
df.plot(kind='box', figsize=(20,9))
```

box plot is used to represent an outlier.

```python
df.plot(kind='box', figsize=(20,9), color={'boxes': 'g', 'whiskers':'r'})
```

```python
df.plot(kind='box', figsize=(20,9), color={'boxes': 'g', 'whiskers':'r'}, vert=False)
```

```python
df.plot(kind='area', figsize=(20,9),stacked=False)
```

```python
df.plot.scatter(x='sepallength', y='petallength')
```

As we see that, when sepallength is increasing petallength is also increasing

```python
df.plot.scatter(x='sepallength', y='petalwidth')
```

In this we see that, the points are scattered so we cannot tell if there is any relationship between these two attributes

```python
df.plot.scatter(x='sepallength', y='petallength',c='sepalwidth',s=df['petalwidth']*200)
```

the third variable will define the color.
size will be defined by petalwidth 

```python
df.plot.hexbin(x='sepallength', y='petalwidth',gridsize=10)
```
